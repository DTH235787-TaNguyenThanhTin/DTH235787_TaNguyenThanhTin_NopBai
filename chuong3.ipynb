```python
# câu 3
from math import sqrt
print("Chương trình Giải Phương trình bậc 2")
a = float(input("Nhập a:"))
b = float(input("Nhập b:"))
c = float(input("Nhập c:"))

if a == 0:
    # bx + c = 0
    if b == 0 and c == 0:
        print("Vô số nghiệm")
    elif b == 0 and c != 0:
        print("Vô nghiệm")
    else:
        x = -c / b
        print("Nghiệm x =", x)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print("Nghiệm kép x1 = x2 =", x)
    else:
        x1 = (-b - sqrt(delta)) / (2*a)
        x2 = (-b + sqrt(delta)) / (2*a)
        print("x1 =", x1)
        print("x2 =", x2)
```

```python
# câu 6:
n = int(input("Nhập số n (0-99): "))
if n < 0 or n > 99:
    print("Số không hợp lệ")
else:
    don_vi = ["", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]
    dac_biet = {15: "mười lăm", 10: "mười", 11: "mười một", 12: "mười hai", 13: "mười ba", 14: "mười bốn"}
    if n < 10:
        if n == 0:
            print("không")
        else:
            print(don_vi[n])
    elif n in dac_biet:
        print(dac_biet[n])
    elif n < 20:
        print("mười", don_vi[n%10])
    else:
        chuc = n // 10
        dv = n % 10
        chuoi = don_vi[chuc] + " mươi"
        if dv == 0:
            print(chuoi)
        elif dv == 1:
            print(chuoi + " mốt")
        elif dv == 5:
            print(chuoi + " lăm")
        else:
            print(chuoi + " " + don_vi[dv])
```

```python
# câu 7
day = int(input("Nhập ngày: "))
month = int(input("Nhập tháng: "))
year = int(input("Nhập năm: "))

def is_leap(y):
    return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)

# Số ngày trong từng tháng
days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
if month == 2 and is_leap(year):
    max_day = 29
elif 1 <= month <= 12:
    max_day = days_in_month[month]
else:
    max_day = 0

if 1 <= month <= 12 and 1 <= day <= max_day:
    day += 1
    if day > max_day:
        day = 1
        month += 1
        if month > 12:
            month = 1
            year += 1
    print(f"Ngày kế tiếp là: {day}/{month}/{year}")
else:
    print("Ngày không hợp lệ!")
```

```python
# câu 8: Máy tính đơn giản với hai số và phép toán
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
op = input("Nhập phép toán (+, -, *, /): ")

if op == '+':
    print(f"{a} + {b} = {a + b}")
elif op == '-':
    print(f"{a} - {b} = {a - b}")
elif op == '*':
    print(f"{a} * {b} = {a * b}")
elif op == '/':
    if b == 0:
        print("Lỗi: Không thể chia cho 0")
    else:
        print(f"{a} / {b} = {a / b}")
else:
    print("Phép toán không hợp lệ!")
```

```python
#câu 18.1: In hình vuông 4x4 rỗng bằng ký hiệu '*'
n = 4
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
#câu 18.2: In ra nửa chéo hình vuông (tam giác vuông cạnh n, cạnh huyền là đường chéo)
n = 4
for i in range(n):
    for j in range(n):
        if j <= i:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
#câu 19: Nhập x, n tính x^(2n+1)/(2n+1)!
x = float(input("Nhập x: "))
n = int(input("Nhập n: "))
mu = 2*n + 1
tu = x ** mu
mau = 1
for i in range(1, mu+1):
    mau *= i
kq = tu / mau
print(f"Giá trị biểu thức x^(2n+1)/(2n+1)! = {kq}")
```
